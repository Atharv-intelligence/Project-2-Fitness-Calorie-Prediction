```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[["Duration"]]
y = df["Calories"]
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
import pandas as pd

model.predict(pd.DataFrame({"Duration": [45]}))
```

```python
model.coef_
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["Duration"], df["Calories"])
plt.xlabel("Duration")
plt.ylabel("Calories")
plt.title("Actual Data")
plt.show()
```

```python
predictions = model.predict(df[["Duration"]])

plt.scatter(df["Duration"], df["Calories"])
plt.plot(df["Duration"], predictions, color="red")
plt.xlabel("Duration")
plt.ylabel("Calories")
plt.title("Model Prediction Line")
plt.show()
```

```python
from sklearn.metrics import r2_score

r2_score(df["Calories"], predictions)
```

“The model achieved an R² score of 0.98, indicating a strong relationship between workout duration and calories burned. However, I understand that in real-world scenarios, performance may vary due to noise and more complex factors.”
